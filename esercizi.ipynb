```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(4,4))
x = np.linspace(-5,5,100)
y = 2*x+1
plt.plot(x, y, '-r', label='y=2x+1')#la label va con la y
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='#1C2843' )
plt.ylabel('y', color='#1C2843')
plt.legend(loc='upper left')#loc serve a dove metterla
plt.grid()
#plt.show()
```

```python
# 100 linearly spaced numbers
x = np.linspace(-5,5,100)
# the function, which is y = ??? here
y = x**3
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)#dividi in quadranti(come assi) la griglia #l'ultimo numero mi dice in quale quadrante lavorare
ax.spines['left'].set_position('center')#asse verticale che con center lo metto al centro del foglio 
ax.spines['bottom'].set_position('center')#asse orizontale che solitamente è in basso con il comando center lo metto al centro del foglio
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')#tacchette sull'asse 
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, '-g')
```

```python
# 100 linearly spaced numbers
x = np.linspace(-np.pi,np.pi,100)
# the function, which is y = sin(x) here
y = np.sin(x)
# setting the axes at the centre
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(2,2, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'or')

# show the plot
z=np.cos(x)
ax = fig.add_subplot(2,2, 3)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,z, 'oy')
##
k = x**3
# setting the axes at the centre
#fig = plt.figure()
ax = fig.add_subplot(2, 2, 2)#dividi in quadranti(come assi) la griglia #l'ultimo numero mi dice in quale quadrante lavorare
ax.spines['left'].set_position('center')#asse verticale che con center lo metto al centro del foglio 
ax.spines['bottom'].set_position('center')#asse orizontale che solitamente è in basso con il comando center lo metto al centro del foglio
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')#tacchette sull'asse 
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,k, '-g')


# show the plot

ax = fig.add_subplot(2, 2 , 4)#in questo modo qui divido il foglio e posiziono le 2 funzioni in posizione 1 e 4 
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'b')
plt.show()
```

```python
# 100 linearly spaced numbers
x = np.linspace(-2,2,100)
# the function, which is y = e^x here
y = np.exp(x)
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'y', label='y=e^x')
plt.legend(loc='upper left')
# show the plot
plt.show()
```

```python
# 100 linearly spaced numbers
x = np.linspace(-4,4,100)
# the functions, which are y = ? and z = ? here
y = np.sin(x)
z = np.cos(x)
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the functions
plt.plot(x,y, 'c', label='y=sin(x)')
plt.plot(x,z, 'm', label='y=cos(x)')
plt.legend(loc='upper left')
# show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
###
##  X label name = Month Number
##  Y label name = Total profit
##
df =pd.read_csv('https://frenzy86.s3.eu-west-2.amazonaws.com/fav/company_sales_data.csv')
```

```python
import calendar

for month_idx in range(1, 13):
    print (calendar.month_name[month_idx])
```

```python
month_word = {
1:'January',
2:'February',
3:'March',
4:'April',
5:'May',
6:'June',
7:'July',
8:'August',
9:'September',
10:'October',
11:'November',
12:'December'
}
```

```python
df['month_number'] = df['month_number'].map(month_word)#metodo map per sostituire il vocabolario creato alla colonna gia presente nella tabella
df
```

```python
# Rename a specific
df.rename(columns={'month_number':'month'},inplace=True)#inplace=true è un altro modo per modificare senza riassegnare il tutto
df
```

```python
## Create a list form column
profitList = df ['total_profit'].tolist()
monthList  = df ['month'].tolist()
```

```python
plt.figure(figsize=(12,8))
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year');

#       color='r', marker='o', markerfacecolor='k', 
#       linestyle='--', linewidth=3);


plt.xlabel('Months')
plt.ylabel('Profit in dollar')

plt.xticks(monthList)
plt.legend(loc='upper left')
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000]);
#plt.xticks([3, 6, 9, 12]);

#plt.show()
```
